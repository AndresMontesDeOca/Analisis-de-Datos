# TP Grupal

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno

pd.set_option('display.max_columns', None)
```

## Data Load

```python
url = 'https://data.insideairbnb.com/argentina/ciudad-aut%C3%B3noma-de-buenos-aires/buenos-aires/2026-01-25/data/listings.csv.gz'
data = pd.read_csv(url)

print(data.info())
display(data.head())
```

## Nulls

```python

# 1. Preparar los datos
null_data = data.isnull().mean().reset_index()
null_data.columns = ['column', 'ratio']
null_data = null_data[null_data['ratio'] > 0].sort_values('ratio', ascending=False)

# 2. Plotear
plt.figure(figsize=(10, 6))
sns.barplot(data=null_data, x='ratio', y='column', palette='Reds_r')

# Añadimos una línea guía para el 50% o 70% (umbral común para borrar columnas)
plt.axvline(0.5, color='black', linestyle='--')
plt.title('Ratio de valores nulos por columna (Solo columnas con nulos)')
plt.xlabel('Proporción de Nulos (0.0 a 1.0)')
plt.show()
```

```python
# Solo columnas con al menos un nulo
cols_con_nulos = data.columns[data.isnull().any()]

# Configuramos el tamaño de la figura antes de llamar a msno
plt.figure(figsize=(12, 7))

# Personalizamos la matriz
# El color (0.2, 0.4, 0.6) es un azul acero elegante
msno.matrix(data[cols_con_nulos],
            figsize=(12, 7),
            color=(0.2, 0.4, 0.6),
            fontsize=12,
            sparkline=True,
            labels=True)

# Agregamos un título profesional con Matplotlib
plt.title('Mapa de Calor de Datos Faltantes', fontsize=20, pad=20)

plt.show()
```
